````python
symbols = ["\u2660", "\u2661", "\u2662", "\u2663"]
height = 20
i = 0
while i < height:
    # 符號
    lsymbol = " "
    msymbol = symbols[i%len(symbols)]
    rsymbol = " "
    # 數目
    lcount = i
    mcount = (2 * height - 1) - 2 * i
    rcount = i
    print(lsymbol * lcount +
       msymbol * mcount +
       rsymbol * rcount)
    i = i + 1
````

````python
s = "asdfsDfgvv   ,, WqrRrwqr. wQr123123rqw5tAfg"
# 英文大小寫, 妳要幫我統計出來s裡面有多少個小寫 多少個大寫
lcase, ucase = 0, 0
i = 0
while i < len(s):
    n = ord(s[i])
    if ord("A") <= n <= ord("Z"):
        ucase = ucase + 1
    elif ord("a") <= n <= ord("z"):
        lcase = lcase + 1
    i = i + 1
print(ucase, lcase)
````

````python
# 妳要幫我統計大小寫字母每一個字母各自有幾個
s = "asdfsDfgvv   ,, WqrRrwqr. wQr123123rqw5tAfg"
ucases = [0] * 26
lcases = [0] * 26
i = 0
while i < len(s):
    n = ord(s[i])
    if ord("A") <= n <= ord("Z"):
        # 65 - 90 -> 0 - 25
        n = n - ord("A")
        ucases[n] = ucases[n] + 1
    elif ord("a") <= n <= ord("z"):
        # 97 - 122 -> 0 - 25
        n = n - ord("a")
        lcases[n] = lcases[n] + 1
    i = i + 1
print(ucases)
print(lcases)

# 印出26行
# A: 5  a: 3
# B: 2  b: 4
# chr(97) -> "a" chr(65) -> "A"
i = 0
while i < len(ucases):
    c1, c2 = chr(i + ord("A")), chr(i + ord("a"))
    n1, n2 = ucases[i], lcases[i]
    print(c1, n1, "/", c2, n2)
    i = i + 1
````

````python
# x的最左邊座標和x的最右邊座標
x_left, x_right = -1, 1
# 分成幾分(100)
part = 1000000
# 每一part的x多長
interval = (x_right - x_left) / part

# 總面積
total = 0
# 積分法: 把每一part的面積(小長方形)加總起來
i = 0
while i < part:
    # 每一part的左x座標: -1 + 0.02 * i
    x = x_left + interval * i
    # 算出這裡的y座標多長: 2 * (1 - x ^ 2) ^ 0.5
    y = 2 * (1 - x ** 2) ** 0.5
    # 每一part面積
    area = interval * y
    # 把每一part累積進去
    total = total + area
    i = i + 1
print("積分法估算面積:", total)
````
