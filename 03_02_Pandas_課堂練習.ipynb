```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/drive/MyDrive/data/vgsales.csv")
data.head()
```

```python
# 找出有缺失值的欄位
data.isnull().sum()
```

```python
# 計算欄位的基本統計數據
data.describe()
```

```python
# 取出 "Publisher" 欄位的資料
data["Publisher"]
```

```python
# 取出 Publisher, Name, Genre 欄位的資料
data[["Publisher","Name","Genre"]]
```

```python
# 計算每筆資料的 NA_Sales,EU_Sales,JP_Sales,Other_Sales 總和

# data["NA_Sales"] + data["EU_Sales"] + data["JP_Sales"] + data["Other_Sales"]

data[["NA_Sales","EU_Sales","Other_Sales"]].sum(axis=1)
```

```python
# 計算每筆資料的 NA_Sales,EU_Sales,JP_Sales,Other_Sales 總和，並存入新的欄位"總和"
data['總和'] = data[["NA_Sales","EU_Sales","Other_Sales"]].sum(axis=1)
data
```

```python
#找出發行商為Nintendo的遊戲名稱

data[data["Publisher"] == "Nintendo"]
```

```python
data[data["Publisher"] == "Nintendo"]["Name"].unique().size
```

```python
#列出所有Wii的遊戲名稱
data[data["Platform"] == "Wii"]["Name"]
```

```python
#計算所有GBA遊戲的銷售總合
data[data["Platform"] == "GBA"]["Global_Sales"].sum()
```

```python
#計算任天堂運動類遊戲的銷售總量

c1 = data['Publisher'] == 'Nintendo'
c2 = data['Genre'] == 'Sports'
data[c1 & c2]['Global_Sales'].sum()

#data[data["Genre"] == "Sports"][data["Publisher"] == "Nintendo"]["Global_Sales"].sum()

data[(data["Genre"] == "Sports") & (data["Publisher"] == "Nintendo")]["Global_Sales"].sum()
```

```python
#列出北美地區銷售量最高的前五名遊戲
# index = data["NA_Sales"].sort_values(ascending=False).index[0:5]
# data.iloc[index]

data.sort_values(by='NA_Sales',ascending=False).head(5)
```

```python
#列出北美與歐洲地區銷售量總和最高的前五名遊戲
#北美+EU

temp_col = data["NA_Sales"] + data["EU_Sales"]
index = temp_col.sort_values(ascending=False).index[0:5]
data.iloc[index]
```

```python
#找出總共有多少種不同的平台
data["Platform"].unique().size
```

```python
#寫程式計算不同平台的總銷售量
for plat in data["Platform"].unique():
  print(plat,round(data[data["Platform"] == plat]["Global_Sales"].sum(),2))
```

```python
# 使用GroupBy來完成上述的工作
data.groupby("Platform")["Global_Sales"].sum()
```
